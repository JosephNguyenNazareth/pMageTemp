```python
file_name = "./case_study/brake_certification/common_terms_tfidf.txt"
file_terms = open(file_name, "r")
terms = file_terms.readlines()
```

```python
tokens = [terms[i].split(",")[0] for i in range(len(terms))]
print(tokens[:20])
```

```python
from nltk.corpus import stopwords
```

```python
def extract_related_words(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words and token.isalnum()]

    # Get related words for each token
    related_words_dict = {}
    for token in filtered_tokens:
        related_terms = related_words.get_related_words_glove(token, related_words.glove_model)
        related_words_dict[token] = related_terms

    related_words_only = {}
    for token, related_words_found in related_words_dict.items():
        related_words_only[token] = []
        for word in related_words_found:
            related_words_only[token].append(word[0])

    return related_words_only
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token not in stop_words and token.isalnum()]

# Get related words for each token
related_words_dict = {}
for token in filtered_tokens:
    related_terms = related_words.get_related_words_glove(token, related_words.glove_model)
    related_words_dict[token] = related_terms
```

```python
related_words_dict
```

```python
related_words_only = {}
for token, related_words_found in related_words_dict.items():
    related_words_only[token] = []
    for word in related_words_found:
        related_words_only[token].append(word[0])
```

```python
related_tokens = related_words_only
```

```python
related_tokens = related_words.extract_related_words(tokens)
```

```python
final_vocab = {}
domain_terms = list(related_tokens.keys())
for key,value in related_tokens.items():
    related_words_2nd_level = extract_related_words(value)
    print(related_words_2nd_level)
    if len(list(related_words_2nd_level.values())) > 0:
        expanded_related_terms = set(np.concatenate(list(related_words_2nd_level.values())))
        expanded_related_terms.update(list(value))
        final_vocab[key] = list(expanded_related_terms.difference(domain_terms))
    else:
        final_vocab[key] = list(domain_terms)
```

```python
final_vocab
```

```python
file = open("./case_study/brake_certification/large_vocab.txt", 'w') 
file.write("domain,general\n")
for term, related_set in final_vocab.items():
    for related_term in related_set:
        file.write(f"{term},{related_term}\n")
file.close()
```
